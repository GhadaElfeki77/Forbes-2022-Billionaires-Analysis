```python
df_gender = df.groupby(["gender"])
```

```python
plt.figure(figsize=(6,4))
df_gender.size().plot(kind="bar", color=["#e9534b", "#4575b4"])
plt.title("Number of Billionaires by Gender (Forbes 2022)", size=14)
plt.xlabel("Gender", size=12)
plt.ylabel("Count", size=12)
plt.savefig('Number of Billionaires by Gender.png')
plt.show()

```

```python
top5_women = df[df["gender"]=="F"].sort_values(by="finalWorth", ascending=False).head(5)
```

```python
sns.barplot(y=top5_women["personName"], x=top5_women["finalWorth"], palette="magma")
plt.title("Top 5 Richest Women in the World (Forbes 2022)")
plt.xlabel("Net Worth (in millions USD)")
plt.ylabel("Name")
plt.savefig('Top 5 Richest Women in the World.png')
plt.show()
```

```python
top_industries_women = (
    df[df["gender"]=="F"]["category"]
    .value_counts()
    .head(5)
)

top_industries_women
```

```python
sns.barplot(x=top_industries_women.values, y=top_industries_women.index, palette="magma")
plt.title("Top 5 Industries for Women Billionaires (Forbes 2022)")
plt.xlabel("Number of Billionaires")
plt.ylabel("Industry")
plt.savefig('Top 5 Industries for Women Billionaires.png')
plt.show()
```

```python
sns.barplot(y = df["personName"][:10], x = df["finalWorth"][:10], palette="magma")
plt.title("Top 10 Richest People in the World (Forbes 2022)", size=16)
plt.xlabel("Net Worth (in millions USD)", size=14)
plt.ylabel("Name", size=14)
plt.savefig('Top 10 Richest People in the World.png')
plt.show()
```

```python
top10_countries = df["country"].value_counts().head(10).index

df_top_countries = df[df["country"].isin(top10_countries)]
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df_top_countries, x="country", hue="gender", palette="magma")
plt.title("Gender Distribution in Top 10 Countries (Forbes 2022)")
plt.xlabel("Country")
plt.ylabel("Number of Billionaires")
plt.xticks(size=10)
plt.legend(title="Gender")
plt.savefig('Gender Distribution in Top 10 Countries.png')
plt.show()
```

```python
df["category"].unique()
```

```python
df["category"]=df["category"].apply(lambda x:x.replace(" ","")).\
    apply(lambda x:x.replace("&","_"))
```

```python
df["category"].unique()
```

```python
df_category = df.groupby("category").size()
```

```python
df_category.head()
```

```python
df_category = df_category.to_frame()
```

```python
df_category = df_category.rename(columns={0:"Count"}).sort_values(by="Count",
                                                                ascending=False)
```

```python
df_category.head()
```

```python
sns.barplot( x = df_category["Count"][:10], y = df_category.index[:10], palette="magma")
plt.title("Top 10 Industries with Most Billionaires (Forbes 2022)", size=16)
plt.xlabel("Number of Billionaires", size=14)
plt.ylabel("Industry", size=14)
plt.savefig('Industries with Most Billionaires.png')
plt.show()
```

```python
industry_gender = (
    df.groupby(["category", "gender"])["personName"]
    .count()
    .reset_index()
    .rename(columns={"personName": "Count"})
)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=industry_gender.sort_values(by="Count", ascending=False),
    x="Count",
    y="category",
    hue="gender",
    palette="magma"
)
plt.title("Gender Distribution by Industry (Forbes 2022)")
plt.xlabel("Number of Billionaires")
plt.ylabel("Industry")
plt.legend(title="Gender")
plt.savefig('Gender Distribution by Industry.png')
plt.show()
```

```python
sns.scatterplot(x=df["age"], y=df["finalWorth"], palette="magma")
plt.title("Relationship between Age and Net Worth (Forbes 2022)")
plt.xlabel("Age")
plt.ylabel("Net Worth (in millions USD)")
plt.savefig('Relationship between Age and Net Worth.png')
plt.show()
```
